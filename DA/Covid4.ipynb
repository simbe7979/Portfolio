```python
# pandas 라이브러리 호출
import pandas as pd
```

```python
# csv형식의 코로나 데이터(주 기준)를 pandas 라이브러리로 읽어오기
state_df = pd.read_csv('C:\\data\\states-covid-data.csv')
state_df
```

```python
# 코로나 데이터에서 필요컬럼(날짜, 주, 확진자 수)만 추출
filtered_state_df = state_df[["submission_date", "state", "tot_cases"]]
filtered_state_df
```

```python
# submission_date 컬럼의 데이터 타입을 datetime으로 변경한 후 그 값을 date 컬럼에 저장
filtered_state_df["date"] = pd.to_datetime(filtered_state_df["submission_date"])
filtered_state_df
```

```python
# date를 기준으로 오름차순으로 정렬
filtered_state_df.sort_values(by = "date", inplace = True)
filtered_state_df
```

```python
# date 컬럼을 인덱스로 설정
filtered_state_df.set_index("date", inplace = True)
filtered_state_df
```

```python
# CA주의 정보만 ca_df에 담기
ca_df = filtered_state_df[filtered_state_df.state == "CA"]
# TX주의 정보만 tx_df에 담기
tx_df = filtered_state_df[filtered_state_df.state == "TX"]
# VA주의 정보만 va_df에 담기
va_df = filtered_state_df[filtered_state_df.state == "VA"]
# IL주의 정보만 il_df에 담기
il_df = filtered_state_df[filtered_state_df.state == "IL"]
# WA주의 정보만 wa_df에 담기
wa_df = filtered_state_df[filtered_state_df.state == "WA"]
```

```python
# 각 주의 확진자 수 컬럼만 추출
ca_total_cases = ca_df["tot_cases"]
tx_total_cases = tx_df["tot_cases"]
va_total_cases = va_df["tot_cases"]
il_total_cases = il_df["tot_cases"]
wa_total_cases = wa_df["tot_cases"]
```

```python
# 각 주의 인구수 조사
ca_population = 39030000
tx_population = 30030000
va_population = 8684000
il_population = 12580000
wa_population = 7786000
```

```python
# CA주의 인구수를 기준으로 TX주와의 인구수 비율 계산
ca_tx_rate = round((ca_population / tx_population), 2)
# CA주의 인구수를 기준으로 VA주와의 인구수 비율 계산
ca_va_rate = round((ca_population / va_population), 2)
# CA주의 인구수를 기준으로 IL주와의 인구수 비율 계산
ca_il_rate = round((ca_population / il_population), 2)
# CA주의 인구수를 기준으로 WA주와의 인구수 비율 계산
ca_wa_rate = round((ca_population / wa_population), 2)
ca_tx_rate, ca_va_rate, ca_il_rate, ca_wa_rate, ca_tx_rate
```

```python
# 시각화용 데이터 생성
final_state_df = pd.DataFrame({
'CA': ca_total_cases,
'TX' : tx_total_cases * ca_tx_rate,   # CA주와의 인구수 차이를 없애기 위해 ca_tx_rate 곱하기
'VA' : va_total_cases * ca_va_rate,   # CA주와의 인구수 차이를 없애기 위해 ca_va_rate 곱하기
'IL' : il_total_cases * ca_il_rate,   # CA주와의 인구수 차이를 없애기 위해 ca_il_rate 곱하기
'WA' : wa_total_cases * ca_wa_rate,   # CA주와의 인구수 차이를 없애기 위해 ca_wa_rate 곱하기
}, index = filtered_state_df.index)
final_state_df
```

```python
# plot.line 함수를 사용하여 시각화
final_state_df.plot.line(rot=45)
```
